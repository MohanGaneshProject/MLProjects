```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
le = LabelEncoder()
norm  = MinMaxScaler()

from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("C:/Users/HP/OneDrive/Desktop/DataAnalysisFiles/Datasets/brain_stroke.csv")
data.head()
```

```python
#Missing values checking
data.isnull().sum()
```

```python
#Initailize the data
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X
```

```python
# Label_Encoding
cat = data.select_dtypes(include='object').columns
cat
```

```python
for i in range(len(cat)):
    X[cat[i]] = le.fit_transform(X[cat[i]])
```

```python
X
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,y_train)
y_log = log.predict(X_test)
accuracy_score(y_test,y_log)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
y_knn = knn.predict(X_test)
accuracy_score(y_test,y_knn)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
y_dtree = dtree.predict(X_test)
accuracy_score(y_test,y_dtree)
```

# RandomForest Classifier


```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
y_clf = clf.predict(X_test)
accuracy_score(y_test,y_clf)
```

# SVM

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
y_svc = svc.predict(X_test)
accuracy_score(y_test,y_svc)
```

# Threshold Values

```python
def select_threshold(model,X_test,threshold):
    return np.where(np.round(model.predict_proba(X_test)[:,0],3)>=threshold,0,1)
```

```python
#Predicting the threshold
from sklearn.metrics import confusion_matrix
for i in np.round(np.arange(0.1,1,0.1),1):
    y_pred = select_threshold(clf,X_test,i)
    print("Threshold value:",i)
    print("Confussion Matrix")
    print(confusion_matrix(y_test,y_pred))
    print("Accuracy Score = ",accuracy_score(y_test,y_pred),"\n")
    
```
